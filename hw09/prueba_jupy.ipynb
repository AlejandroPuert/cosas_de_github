```python
def rungekutta4(dt,t,X,Xprima):
    assert callable(Xprima) 
    k1 = Xprima(X,t)*dt
    k2 = Xprima(X+k1/2,t+dt/2)*dt
    k3 = Xprima(X+k2/2,t+dt/2)*dt
    k4 = Xprima(X+k3,t+dt)*dt
    return X+(k1+2*k2+2*k3+k4)/6
def Alpha(X,t):
    vel = X[1]
    X_prima = [vel,-9.8]
    X_prima = np.array(X_prima)
    return X_prima
```

```python
X=np.array([0,20])
t = 0
dt = 0.1
Xlist = [X]
tlist = [0]
xnew = [0]
while xnew[0]>=0:
    t+=dt
    xnew = rungekutta4(dt,t,Xlist[-1],Alpha)
    Xlist.append(xnew)
    tlist.append(t)
Xlist = np.array(Xlist)
```

```python
plt.plot(tlist,Xlist[:,0])
plt.grid()
```
